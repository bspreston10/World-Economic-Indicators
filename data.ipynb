```python
import pandas as pd
import numpy as np  
```

```python
world_bank = pd.read_excel('WorldBank.xlsx', sheet_name='Development Indicators').rename({'Country Name': 'Country'}, axis=1)

world_bank['Population(M)'] = (world_bank['GDP (USD)'] / world_bank['GDP per capita (USD)'] / 1_000_000)


world_bank.head()
```

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi = pd.read_csv('HDI.csv')

w_hdi_2014 = world_bank.query('Year == 2014').merge(hdi[['iso3', 'hdi_2014']], how='left', left_on='Country Code', right_on='iso3')

w_hdi_2014.head()
```

```python
w_hdi_2014.describe()
```

```python
gdp_pivot = world_bank.pivot_table(index='Year', columns='Region', values='GDP (USD)', aggfunc='sum')

gdp_pivot.head()
```

```python
pop_pivot = world_bank.pivot_table(index='Year', columns='Region', values='Population(M)', aggfunc='sum')

pop_pivot.head()
```

```python
w_hdi_by_region = w_hdi_2014.groupby('Region').agg({'hdi_2014': 'mean'}).sort_values(by='hdi_2014', ascending=False)
w_hdi_by_region.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

fig, ax = plt.subplots()

ax.stackplot(
    gdp_pivot.index, 
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend(loc='upper left')
ax.set_title('Exponential Growth of GDP by Region')
ax.set_xlabel('Year')
ax.set_ylabel('GDP (Trillions USD)')
```

```python
fig, ax = plt.subplots()

ax.stackplot(
    pop_pivot.index, 
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=pop_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend(loc='upper left')
ax.set_title('Linear Growth of Population by Region')
ax.set_xlabel('Year')
ax.set_ylabel('Population (Billions)')
```

```python
minize = min(w_hdi_2014['Population(M)'])
maxize = max(w_hdi_2014['Population(M)'])

sns.scatterplot(
    data=w_hdi_2014, 
    x='Life expectancy at birth (years)', 
    y='GDP per capita (USD)', 
    size='Population(M)', 
    sizes=(minize, maxize),
    hue='Region'
).set(title='Increase in Life Expectancy as GDP Increases', yscale = 'log')
```

```python
w_hdi_by_region.plot.bar(legend=False, title='HDI by Region', ylabel='HDI')
```

```python
sns.scatterplot(
    data=w_hdi_2014.query("Country != 'Iceland'"),
    x='Electric power consumption (kWh per capita)',
    y='GDP per capita (USD)',
    hue='hdi_2014',
    palette= 'coolwarm_r'
).set(title='Electric Power Consumption vs GDP per Capita')
```

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(8, 12))

gs = GridSpec(12, 6, figure=fig)

stack1_list = ['#66C2A5', '#FC8D62', '#8DA0CB', '#E78AC3', '#A6D854', '#FFD92F', '#B3B3B3']
stack2_list = ['#66C2A5', '#A6D854', '#B3B3B3', '#FC8D62', '#E78AC3', '#8DA0CB', '#FFD92F']
bar_list = ['#8DA0CB', '#FC8D62', '#E78AC3', '#FFD92F', '#66C2A5', '#A6D854', '#B3B3B3']
bubble_list = ['#A6D854', '#FC8D62', '#FFD92F', '#66C2A5', '#B3B3B3', '#E78AC3', '#8DA0CB']

fig.suptitle('Report: Global Economic Development from 1960-2018', fontsize=16, fontweight='bold', x=0.6, y=0.93)  


ax1 = fig.add_subplot(gs[0:4, 0:3])

ax1.stackplot(
    gdp_pivot.index, 
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_list
)

ax1.set_title('Exponential Growth of GDP by Region')
ax1.set_ylabel('GDP (Trillions USD)')

fig.text(.92, .63, 
'''Economic expansion in the 20th and early 
21st centuries has been enormous and has 
outpaced population growth. In 1960, Europe
and North America produced 75% of GDP on 
20% of the population.

By 2018, East Asia had surpassed both
in economic output largely due to 
the rise of Japan, South Korea, and China.

In the coming decades, economists expect
South Asia and Sub-Saharan Africa to lead
the world in global growth.''', fontsize=8)

ax2 = fig.add_subplot(gs[0:4, 3:6])

ax2.stackplot(
    pop_pivot.index, 
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=pop_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack2_list
)

ax2.legend(loc='upper left')
ax2.set_title('Linear Growth of Population by Region')
ax2.set_ylabel('Population (Billions)')
ax2.legend(bbox_to_anchor=(1.75, 1), fontsize=8)

ax3 = fig.add_subplot(gs[4:8, :])

minize = min(w_hdi_2014['Population(M)'])
maxize = max(w_hdi_2014['Population(M)'])

sns.scatterplot(
    data=w_hdi_2014, 
    x='Life expectancy at birth (years)', 
    y='GDP per capita (USD)', 
    size='Population(M)', 
    sizes=(minize, maxize),
    hue='Region',
    palette=bubble_list,
    ax=ax3
).set(title='Increase in Life Expectancy as GDP Increases', yscale = 'log')

handles, labels = ax3.get_legend_handles_labels()
entries_to_skip = len(w_hdi_2014["Region"].unique())+1



ax3.legend(handles[entries_to_skip:], labels[entries_to_skip:], bbox_to_anchor=(0.8, 0.9), borderaxespad=0., frameon=False, ncol=6, fontsize=8)

fig.text(.92, .42,
'''The wealthy nations of the world
enjoy high GDP per capita as well
as long life spans, but make up 
a relatively small share of the
global population.

If growth can continue in the developing
world, humanity will be vastly wealthier and 
healthier.''', fontsize=8)

ax4 = fig.add_subplot(gs[8:12, 0:3])

w_hdi_by_region.plot.bar(ax=ax4, legend=False, title='HDI by Region', ylabel= 'HDI', color= bar_list)

ax5 = fig.add_subplot(gs[8:12, 3:6])

sns.scatterplot(
    data=w_hdi_2014.query("Country != 'Iceland'"),
    x='Electric power consumption (kWh per capita)',
    y='GDP per capita (USD)',
    hue='hdi_2014',
    palette= 'coolwarm_r',
    ax=ax5
).set(title='Electric Power Consumption vs GDP per Capita')

fig.text(.92, .15,
'''HDI, short for Human Development Index, 
attempts to measure the quality of life
in countries. Life Expectancy, Education,
and GDP per capita are the factors that
are considered.

Economic growth in the developing world
in the 21st century should help other regions
catch up to North America and Europe.

One factor not in HDI, but essential in 
a country's development is electricity consumption.
Electricity is a key driver of economic growth, and 
developing countries need to increase their
investment in power infrastructure to ensure growth.''', fontsize=8)

fig.subplots_adjust(wspace=4, hspace=4)
```
